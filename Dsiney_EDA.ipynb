### Relationships Between Total Gross and Other Fields

### Genre

```python
disney.groupby("genre").describe()
```

```python
genre_group = disney.groupby("genre")
```

```python
for i in set(disney["genre"]):
    print("-"*16, i, "-"*16)
    grp = disney.where(disney["genre"]==i).dropna()
    print("\n\n----MAX----\n")
    print(grp.max())
    print("\n\n----MIN----\n")
    print(grp.min())
    print("\n\n")
```

```python
for i in set(disney["genre"]):
#     print("-"*16, i, "-"*16)
#     fig=plt.figure()
#     ax=fig.add_axes([0,0,1,1])
    grp = disney.where(disney["genre"]==i).dropna()
    grp.plot(x="release_date", y="total_gross", title="Release Date Vs Total Gross For %s Genre"%(i))
#    ax.set_title("Release Date Vs Total Gross For %s Genre"%(i))
#    ax.plot(x=grp["release_date"], y=grp["total_gross"])
```

```python
for i in set(disney["genre"]):
    grp = disney.where(disney["genre"]==i).dropna()
    grp.plot(kind="scatter", x="release_date", y="total_gross", title="Release Date Vs Total Gross For %s Genre"%(i))
```
